```python
def add_log(logs):  # Accept logs as a parameter to pass them to the function.
    s = input("Enter log (format: time:level:user_id:message): ").split(':', 3)
    if len(s) != 4:
        print("Invalid log format.")
        return logs

    log = {
        'time': s[0],
        'level': s[1],
        'user_id': int(s[2]),
        'message': s[3]
    }
    logs.append(log)
    return logs

def get_user_logs(logs, user_id):  # Returns a list of logs generated by the given user.
    user_logs = []
    for log in logs:
        if log['user_id'] == user_id:
            user_logs.append(log)
    return user_logs

def count_levels(logs):  # Returns a dictionary showing the frequency of each log level.
    levels = {}
    for log in logs:
        for level in ["Info", "Warning", "Error", "Debug", "Critical"]:  # Add any levels you expect
            if log['level'] == level:
                levels[level] = levels.get(level, 0) + 1
                break  # Assumes one level per log entry
    return levels

def filter_logs(logs, keyword):  # Returns all logs whose message contains the given keyword (case-insensitive).
    keyword_logs = []
    for log in logs:
        if keyword.lower() in log.lower():  # Corrected logic to search within 'message'
            keyword_logs.append(log)
    return keyword_logs

def get_recent_logs(logs, n=2):  # Returns the most recent 'n' logs in the system.
    return logs[-n:] if len(logs) >= n else logs

if __name__ == '__main__':
    logs = []
    logs = add_log(logs)  # Pass logs to the function
    user_logs = get_user_logs(logs, 1)
    print(user_logs)

    count = count_levels(logs)
    print(count)

    filtered = filter_logs(logs, "Start")
    print(filtered)

    recent_logs = get_recent_logs(logs, 3)
    print(recent_logs)
```

```python
def add_log(logs):
    s = input("Enter log (format: yyyy-mm-dd HH:MM:SS:Level:UserID:Message): ").split(':')

    if len(s) < 5:
        print("Invalid log format.")
        return logs

    time = ':'.join(s[:3])
    level = s[3]
    user_id = int(s[4])
    message = ':'.join(s[5:])

    log = {
        'time': time,
        'level': level,
        'user_id': user_id,
        'message': message
    }
    logs.append(log)
    return logs

def get_user_logs(logs, user_id):
    user_logs = []
    for log in logs:
        if log['user_id'] == user_id:
            user_logs.append(log)
    return user_logs

def count_levels(logs):
    levels = {}
    for log in logs:
        level = log['level']
        levels[level] = levels.get(level, 0) + 1
    return levels

def filter_logs(logs, keyword):
    keyword_logs = []
    for log in logs:
        if keyword.lower() in log['message'].lower():
            keyword_logs.append(log)
    return keyword_logs

def get_recent_logs(logs, n=2):
    return logs[-n:] if len(logs) >= n else logs

if __name__ == '__main__':
    logs = []
    logs = add_log(logs)

    user_logs = get_user_logs(logs, 1)
    print(user_logs)

    count = count_levels(logs)
    print(count)

    filtered = filter_logs(logs, "Start")
    print(filtered)

    recent_cap = get_recent_logs(logs, 1)
    print(recent_cap)
```
